**2.  Ahora sí, creamos la función**

```python
def primera_funcion(base):
    
    ## Importamos la base de datos
    consumo_comedor = pd.read_excel('../../../data/consumo_comedor.xls')
    
    ## Primera base requerida: cantidad total de menúsy montos totales por concepto
    b1 = consumo_comedor.groupby( ['Concepto'] )[['cantidad', 'monto Total']].sum()
    
    ## Segunda base requerida: usuarios con códigos raros
    b2 = consumo_comedor
    b2 = b2[ (b2['usuario'] < 1000) | (b2['usuario'] > 9999) ]
    b2f = b2[ ['usuario' , 'Concepto' , 'cantidad' , 'monto Total'] ]
    b2f
    
    ## Tercera base requerida: cantidad de consumo por usuario, según tipo de consumo
    b3 = consumo_comedor.groupby( ['usuario' , 'Concepto'] )[['cantidad', 'monto Total']].sum()
    
    if base == 'base1':
        return b1
    elif base == 'base2':
        return b2f
    elif base == 'base3':
        return b3
```

**3. Comprobamos que la función funciona adecuadamente**
