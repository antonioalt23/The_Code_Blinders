COVERTIRLO A FUNCIÓN

```python
def función(consumo) : 
      
    consumo_comedor = pd.read_excel('../../../data/consumo_comedor.xls')
    
    data_frame_1=consumo_comedor.groupby(['Concepto'])[['cantidad','monto Total']].sum()
    
    data2= consumo_comedor.loc[ consumo_comedor.usuario.notnull(), :]
    filtro_raros = data2['usuario'].astype('string').str.len() != 4
    data3 = data2.loc[filtro_raros, : ]
    
    date_frame_3=consumo_comedor.groupby( ['usuario' , 'Concepto'] )[['cantidad', 'monto Total']].sum()
    
    if consumo == 'consumo1' :
    
        return  data_frame_1
    
    elif consumo == 'consumo2' :
    
        return data3
    
    elif consumo == 'consumo3' :
    
        return date_frame_3
```

```python
función('consumo1')
```

```python
función('consumo2').loc[filtro_raros, : ]
```

```python
función('consumo3')
```
