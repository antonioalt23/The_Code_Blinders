```python
def vamo_a_comer(comida):
    consumo_comedor = pd.read_excel('../../../data/consumo_comedor.xls')
    
    #Juntamos y sumamos las cantidades totales según su concepto
    concepto = consumo_comedor.groupby( ['Concepto'] )[['cantidad', 'monto Total']].sum()
    #AQUI VA EL ACUMULADO DE CONSUMO
    consumo = consumo_comedor.groupby( ['usuario' , 'Concepto'] )[['cantidad', 'monto Total']].sum()
    # AQUI ENCONTRAREMOS A LOS USUARIOS CON CÓDIGOS INUSUALES 
    códigos_inusuales = consumo_comedor
    códigos_inusuales = códigos_inusuales[ (códigos_inusuales['usuario'] < 1000) | (códigos_inusuales['usuario'] > 9999) ]
    ci = códigos_inusuales[ ['usuario' , 'Concepto' , 'cantidad' , 'monto Total'] ]
    ci
    
    
    if comida == 'comida1':
        return concepto
    
    elif comida == 'comida2':
        return consumo
    
    elif comida == 'comida3':
        return ci
    
    else :
        print('No te rindas')
     
    
```
