La siguiente línea muestra dicha función desarrollada.

```python
def function_comedor (data):
    
    # Cargamos la data
    data_comedor = pd.read_excel('../../../data/consumo_comedor.xls')
    
    # 1. Data frame de la totalidad de consumo y el monto total
    #    por tipo de pedido MENÚ ECONÓMICO y DESAYUNO.
    
    data_consumo_comedor = data_comedor.groupby( [ 'Concepto' ] )[["cantidad", "monto Total"]].sum()
    
    # 2. Data frame con aquellos usuarios que no cumplen con tener
    #    un código de usuario de tamaño 4.
    
    data_comedor_usuario_rare = data_comedor[ (data_comedor['usuario'] <= 999 ) | (data_comedor['usuario'] >= 10000 ) ]
    data_comedor_usuario_rare = data_comedor_usuario_rare[["usuario"]]
    
    # 3. Data frame con la cantidad de consumo por usuario dividido en el tipo de consumo.
    
    usuario_dividido1 = data_comedor.groupby( ["Concepto" , "usuario"] )[["cantidad"]].sum() 
    usuario_dividido1
    
    if data == 'DF1':
        return data_consumo_comedor
    elif data == 'DF2':
        return data_comedor_usuario_rare
    elif data == 'DF3':
        return usuario_dividido1

```
