### *DataFrame*

Un DataFrame es una colección de múltiples series.Se puede considerar a los DataFrame como una matriz bidimensional, donde cada fila es un punto de datos separados y una columna es una caracteristica de los datos.Cada una de las filas están etiquetadas con un índice similar al de las Series y el de las columnas están etiquetadas en las columnas de los atributos.Hay muchas formas de iniciar un DataFrame.Una de las más usadas es pasando un diccionario como los datos de un DataFrame .La claves del diccionario se convertirán en etiquetas en las columnas y los valores se convertirán en la serie asociada con dad etiqueta. 

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
# crear dataframe
alumnos = ["Alejandro","Pedro","Ramiro","Axel","Juan"]
calculo = [15,16,10,12,13]
ingles = [13,9,16,14,17]
arte = [12,16,15,19,10]

# crear el diccionario a partir de estas listas
notas = {"alumnos":alumnos, "calculo":calculo, "ingles":ingles, "arte":arte}
```

```python
notas
```

```python
registro = pd.DataFrame(notas)
registro
```

```python
# otra forma 
names =  ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr = [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

reg_cars = {"country":names, "drives_right":dr, "cars_per_cap":cpc}
```

```python
cars = pd.DataFrame(reg_cars)
cars
```

```python
data = {'state': ['Lima', 'Piura', 'Tumbes', 'Cuzco', 'Ica', 'Puno'],
        'year': [2000, 2001, 2002, 2001, 2002, 2003],
        'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
frame = pd.DataFrame(data)
frame
```

```python
# cambiar el orden de las columnas
state_pop = pd.DataFrame(data, columns = ["year","state","pop"])
state_pop
```

```python
# agregar una columna y la indexación
frame2 = pd.DataFrame(data, index=['one', 'two', 'three', 'four','five', 'six'], columns=["year", 'state', 'pop', 'debt'])
frame2
```
