## 0. ¿A qué sexo pertenecen el mayor número de fallecidos?

```python
muertos_sexo=muertos_covid.groupby( [ 'SEXO' ] )[['UUID']].count()
mujeres=muertos_sexo.iloc[2,0]
print(f'El número de mujeres fallecidas es: {mujeres}')
hombres=muertos_sexo.iloc[3,0]
print(f'El número de hombres fallecidos es: {hombres}\n'  )

if hombres>mujeres:
    print('Por lo tanto hay mayor número de hombres fallecidos')
else:
    print('Por lo tanto hay mayor número de mujeres fallecidas')
```

## 1. Dentro de los fallecidos de entre 30 y 60 años, ¿Qué sexo es el que cuenta con mayor número de fallecidos ?

```python
fallecidos_30_60 = muertos_covid.loc[ ( muertos_covid[ "EDAD_DECLARADA" ] >= 30 ) & ( muertos_covid[ "EDAD_DECLARADA" ] <= 60 )]
sexo_30_60=fallecidos_30_60.groupby( [ 'SEXO' ] )[['UUID']].count()

mujeres_30_60=sexo_30_60.iloc[2,0]
print(f'El número de mujeres fallecidas entre 30 y 60 años es: {mujeres}')
hombres_30_60=sexo_30_60.iloc[3,0]
print(f'El número de hombres fallecidos entre 30 y 60 años es: {hombres}\n'  )

if hombres_30_60>mujeres_30_60:
    print('Por lo tanto hay mayor número de hombres entre 30 y 60 años fallecidos')
else:
    print('Por lo tanto hay mayor número de mujeres entre 30 y 60 años fallecidas')
```

## 2. ¿Cuántas observaciones no cuentan con información de fecha de nacimiento?

```python
NanMuertos=muertos_covid['FECHA_NAC'].isnull().sum()
print(f'Existen {NanMuertos} observaciones que no cuentan con información sobre la fecha de nacimiento.')
```

## 6. Genera una tabla donde se encuentre el porcentaje de muertos por cada distrito con respecto a la provincia (llamen a la columna distr_prov) que pertenecen. Este dataframe debe contener las siguientes columnas: DEPARTAMENTO, PROVINCIA, DISTRITO y distr_prov.

```python
database=pd.DataFrame(muertos_covid, columns=["DEPARTAMENTO", "PROVINCIA", "DISTRITO"])
database['distr_prov']=0
count_provincia=database.groupby( [ 'DEPARTAMENTO', 'PROVINCIA' ] ).count()
count_distrito=database.groupby( [ 'DEPARTAMENTO',  'PROVINCIA' ,'DISTRITO' ] ).count()
resultado=count_distrito/count_provincia*100
resultado
```

## 7. Genera una tabla donde se encuentre el porcentaje de muertos por cada provincia con respecto al departamento que pertenecen (llamen a la columna prov_dpto). Este dataframe debe contener las siguientes columnas: DEPARTAMENTO, PROVINCIA y prov_dpto.

```python
database2=pd.DataFrame(muertos_covid, columns=["DEPARTAMENTO", "PROVINCIA"])
database2['prov_dpto']=0
count_departamento=database2.groupby( [ 'DEPARTAMENTO'] ).count()
count_provincia=database2.groupby( [ 'DEPARTAMENTO',  'PROVINCIA' ] ).count()
resultado2=count_provincia/count_departamento*100
resultado2
```

## 8. Genera una tabla donde se encuentre el porcentaje de muertos por cada distrito con respecto al departamento que pertenecen (llamen a la columna distr_dpto). Este dataframe debe contener las siguientes columnas: DEPARTAMENTO, PROVINCIA, DISTRITO, distr_prov, prov_dpto y distr_dpto.

```python
database3=pd.DataFrame(muertos_covid, columns=["DEPARTAMENTO", "PROVINCIA","DISTRITO" ])
database3['distr_dpto']=0
dep_count=database3.groupby( [ 'DEPARTAMENTO'] ).count()
distr_count=database3.groupby( [ 'DEPARTAMENTO',  'PROVINCIA','DISTRITO'  ] ).count()
resultado3=distr_count/dep_count*100
resultado3
```

## 9. Exportar la data de la pregunta 8 en un csv que se guarde en la carpeta save_data. Los datos nuemricos deben ser redondeados en 2 decimales.


```python
import pandas as pd
from pandas import ExcelWriter
name="Tarea3.xlsx" 
writer = ExcelWriter(name)
resultado3.to_excel(writer, 'Hoja de datos', index=False)
writer.save()
```
