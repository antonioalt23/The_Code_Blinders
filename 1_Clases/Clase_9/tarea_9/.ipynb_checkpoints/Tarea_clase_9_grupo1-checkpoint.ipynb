### $\color{blue}{\text{1. }}$ Dentro de los fallecidos de entre 30 y 60 años, ¿Qué sexo es el que cuenta con mayor número de fallecidos?

```python
muertos_covid.groupby( [ 'EDAD_DECLARADA' ] )[['SEXO']].count()
```

```python
# Definimos la variable "adulto" que corresponde al grupo etario entre 30 a 60 años.
adulto = muertos_covid[ (muertos_covid["EDAD_DECLARADA"] >= 30) & (muertos_covid["EDAD_DECLARADA"] <= 60)]

# Filtramos la variable adulto por SEXO
adulto.groupby( [ 'SEXO' ] )[['EDAD_DECLARADA']].count()
```

```python
# RESPUESTA: Dentro de los fallecidos entre 30 y 60 años,
#            el sexo con mayor número de fallecidos es el "MASCULINO" con un total de 10459 fallecidos
# Sería interesante ver la proporción que representan del total.
```

### $\color{blue}{\text{2. }}$ ¿Cuántas observaciones no cuentan con información de fecha de nacimiento?

```python
# Método 1
#----------

# Una visión general para saber si la variable "FECHA_NAC" no cuentan con información es a través del siguiente comando

muertos_covid.info()
```

```python
# Los resultados previos nos muestran que la variable "FECHA_NAC" tiene 31747 valores no nulos.
# Esto quiere decir que el complemento para el total de observaciones (47089) corresponde a los valores nulos.
# Entonces, con la simple aritmética podemos saber las observaciones sin información para dicha variable.
# A saber: 
total_observaciones = 47089

valores_no_nulos = 31747

fecha_nac_sin_info = total_observaciones - valores_no_nulos
print(fecha_nac_sin_info)
# Este resultado coincide con el resultado anterior, por lo que,
# podemos decir que la función info() me otorga los valores sin informació,
# (que en este caso corresponden a la variable "FEHCA_NAC").
```

```python
# Método 2
#----------

# Usaremos la función isnull()

muertos_covid['fecha_nac_sin_info2'] = muertos_covid['FECHA_NAC'].isnull()
FECHA_NAC_sin_info2 = muertos_covid[ (muertos_covid["fecha_nac_sin_info2"] == True)]
print(len(FECHA_NAC_sin_info2))

# Este resultado también coincide con los resultados previos
```
