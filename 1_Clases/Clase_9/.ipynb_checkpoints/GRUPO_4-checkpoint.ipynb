### Relative Path ###
Escribe el relative path que ubica al archivo fallecidos_covid.csv y asiganle a la varible relative_path

```python
import pandas as pd
import numpy as np
import chardet
```

```python
#Gettting the code of strings

rawdata = open(r'../../data/fallecidos_covid.csv', 'rb').read()
result = chardet.detect(rawdata)
charenc = result['encoding']
```

```python
relative_path = '../../data/fallecidos_covid.csv'
muertos_covid = pd.read_csv( relative_path,   delimiter = ';',  encoding = charenc )
muertos_covid.head(5)
```

## Anlisis de la data fallecidos covid

**0. A qu sexo pertenecen el mayor nmero de fallecidos?**

```python
## Primero quiero confirmar que no hay missings en la variable SEXO
muertos_covid['SEXO'].isnull().values.any()
# No hay
```

```python
muertos_covid[ 'SEXO' ].unique()
# Hay dos valores (10 y 11) que parecen errores de registro
```

```python
# Ahora s realizamos el conteo
muertos_covid.groupby( [ 'SEXO' ] ).count()
# Hay 10 personas no clasificadas (10 y 11)
```

**1. Dentro de los fallecidos de entre 30 y 60 aos, Qu sexo es el que cuenta con mayor nmero de fallecidos ?**

```python
# Tomando en cuenta la distribucin de la base por sexo y el rango dedad especificado, es muy probable que hayan ms hombres
mc_30_60 = muertos_covid[ (muertos_covid["EDAD_DECLARADA"] >= 30) & (muertos_covid["EDAD_DECLARADA"] <= 60)]
mc_30_60.groupby(['SEXO']).count()
# Efectivamente, hay ms hombres
```

**2. Cuntas observaciones no cuentan con informacin de fecha de nacimiento?**

```python
muertos_covid['FECHA_NAC_null'] = muertos_covid['FECHA_NAC'].isnull()
fn_null = muertos_covid[ (muertos_covid["FECHA_NAC_null"] == True)]
len(fn_null)
```

**3. Halla el total de fallecidos por departamento.**

```python
# Sabiendo que la variable UUID no tiene missings, podemos usarla para el conteo a cualquier nivel
muertos_covid.groupby( ['DEPARTAMENTO'] ).count()
```

**4. Halla el total de fallecidos por provincia.**

```python
# Sabiendo que la variable UUID no tiene missings, podemos usarla para el conteo a cualquier nivel
muertos_covid.groupby( ['DEPARTAMENTO' , 'PROVINCIA'] ).count()
```

**5. Halla el total de fallecidos por distrito.**

```python
# Sabiendo que la variable UUID no tiene missings, podemos usarla para el conteo a cualquier nivel
muertos_covid.groupby( ['DEPARTAMENTO' , 'PROVINCIA' , "DISTRITO"] ).count()
```
