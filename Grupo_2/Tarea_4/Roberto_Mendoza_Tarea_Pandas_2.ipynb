### Halla el total de fallecidos por departamento.

```python
base = pd.DataFrame(muertos_covid.groupby( ["DEPARTAMENTO"] ).count())
base.rename(columns = {'UUID':'Muertos por departamento'}, inplace = True)
muertos_departamento=base[['Muertos por departamento']]
muertos_departamento
```

### Halla el total de fallecidos por provincia.

```python
base = pd.DataFrame(muertos_covid.groupby( ["PROVINCIA"] ).count())
base.rename(columns = {'UUID':'Muertos por provincia'}, inplace = True)
muertos_provincia=base[['Muertos por provincia']]
muertos_provincia
```

### Halla el total de fallecidos por distrito.

```python
base= pd.DataFrame(muertos_covid.groupby( ["DISTRITO"] ).count())
base.rename(columns = {'UUID':'Muertos por distrito'}, inplace = True)
muertos_distrito=base[['Muertos por distrito']]
muertos_distrito
```

### Genera una tabla donde se encuentre el porcentaje de muertos por cada distrito con respecto a la provincia (llamen a la columna distr_prov) que pertenecen. Este dataframe debe contener las siguientes columnas: DEPARTAMENTO, PROVINCIA, DISTRITO y distr_prov.

```python
base_1=muertos_covid.merge( muertos_departamento , how = 'left', on = 'DEPARTAMENTO' , indicator = False )
base_2=base_1.merge( muertos_provincia , how = 'left', on = 'PROVINCIA' , indicator = False )
base_total=base_2.merge( muertos_distrito , how = 'left', on = 'DISTRITO' , indicator = False )
```

```python
base_total['distr_prov']=(base_total['Muertos por distrito']/base_total['Muertos por provincia'])*100
base_total['prov_dpto']=(base_total['Muertos por provincia']/base_total['Muertos por departamento'])*100
base_total['distr_dpto']=(base_total['Muertos por provincia']/base_total['Muertos por departamento'])*100
Tabla_1 = base_total[['DEPARTAMENTO','PROVINCIA','DISTRITO','distr_prov']].round(2)
Tabla_1
```

### Genera una tabla donde se encuentre el porcentaje de muertos por cada provincia con respecto al departamento que pertenecen (llamen a la columna prov_dpto). Este dataframe debe contener las siguientes columnas: DEPARTAMENTO, PROVINCIA y prov_dpto.

```python
Tabla_2 = base_total[['DEPARTAMENTO','PROVINCIA','prov_dpto']].round(2)
Tabla_2
```
