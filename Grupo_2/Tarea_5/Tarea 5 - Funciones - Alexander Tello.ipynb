```python
import numpy as np
import pandas as pd 
from pandas import DataFrame,Series
```

```python
consumo_comedor = pd.read_excel(r"The_Code_Blinders/data/consumo_comedor.xls").drop(['Unnamed: 1'],axis = 1)
consumo_comedor.head()
```

```python
def consumo_comedor (retorno = 'Concepto', modo = 'DataFrame'):
    x = pd.read_excel(r"The_Code_Blinders/data/consumo_comedor.xls").drop(['Unnamed: 1'],axis = 1)
    
    if retorno == "Concepto":
        tabla = x.groupby(['Concepto'])['cantidad','monto Total'].sum()
        tabla1 = pd.DataFrame(tabla.agg(['sum']).values, columns = ['cantidad', 'monto Total'],index=['TOTAL'])
        tabla = tabla.append(tabla1)
        if modo == 'DataFrame':
            return tabla
        elif modo == 'Dic':
            return tabla.to_dict('dict')
        else:
            print('Error, inserta el MODO correcto')
    elif retorno == "Usuarios_Raros":
        tabla2 = x[(x['usuario']<1000) | (x['usuario']>9999)]
        tabla2 = tabla2[['usuario','cantidad','Concepto','monto Total']]
        if modo == 'DataFrame':
            return tabla2
        elif modo == 'Dic':
            return tabla2.to_dict('dict')
        else:
            print('Error, inserta el MODO correcto')
    elif retorno == "Por_Usuario":
        tabla3 = x.groupby(['usuario','Concepto'])['cantidad','monto Total'].sum().reset_index()
        if modo == 'DataFrame':
            return tabla3
        elif modo == 'Dic':
            return tabla3.to_dict('dict')
        else:
            print('Error, inserta el MODO correcto')
    else:
        print('Error, inserta el NOMBRE correcto')
    
```
