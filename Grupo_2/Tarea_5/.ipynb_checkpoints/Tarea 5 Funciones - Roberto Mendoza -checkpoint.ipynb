#### Devolver un tipo particular de data frame requiere ingresar el nombre adeuado
#### 1. Consumo : cuadro que muestra el monto por consumo y total
#### 1. Codigos : cuadro que muestra los clientes con códigos de usuario mayor a cuatro
#### 1. Consumo_cliente :cuadro que muestra el tipo consumo por cada de cliente 

```python
def cuadros( base_datos, tipo):
    
    filtro_1 = consumo["Concepto"] == "MENU ECONOMICO"
    filtro_2 = consumo["Concepto"] == "DESAYUNO"
    ME = consumo["monto Total"].loc[filtro_1].sum()
    DE = consumo["monto Total"].loc[filtro_2].sum()
    TO = consumo["monto Total"].loc[filtro_1 | filtro_2].sum()
    data1 = {'CONSUMO': ['MENU ECONOMICO', 'DESAYUNO','TOTAL'],
        'MONTO': [ME,DE,TO]}
    cuadro_1 = pd.DataFrame(data1)
    
    j = 0
    consumo["codigos"] = ""
    for i in consumo["usuario"]:
        leng = len(str(i)) 
    
        if leng == 6: 
            consumo.loc[[j], ["codigos"]] = "completo"
        else:
            consumo.loc[[j], ["codigos"]] = "Incompleto"         
            
        j = j + 1
   
     
    filtro_3 = consumo["codigos"] == "Incompleto" 
    cuadro_2 = consumo[filtro_3].groupby([ 'usuario' ])[['monto Total']].sum()

    usuario_1 = consumo[filtro_1].groupby([ 'usuario' ])[['monto Total']].sum()
    usuario_2 = consumo[filtro_2].groupby([ 'usuario' ])[['monto Total']].sum()
    cuadro_3 = usuario_1.merge( usuario_2 , how = 'outer', on = [ 'usuario'])
    cuadro_3.rename( columns = {'monto Total_x':'Monto Total Menu Economico', 'monto Total_y':'Monto Total Desayuno'}, inplace = True)
    
    if tipo == "Consumo":
            return cuadro_1
    if tipo == "Codigos":
            return cuadro_2
    if tipo == "Consumo_cliente":
            return cuadro_3
    else:
            print("Ingresar la solicitud adecuada")
```

```python
cuadros(consumo, "Codigos")
```
